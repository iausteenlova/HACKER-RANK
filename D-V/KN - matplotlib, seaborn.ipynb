```python
import matplotlib.pyplot as plt
%matplotlib inline
```

> Simple Scatter plot

```python
x = np.arange(1,20,dtype=int)
y = np.arange(11,30,dtype=int)
```

```python
plt.scatter(x,y,marker='8',c='r')
plt.xlabel('X')
plt.xlabel('Y')
plt.title('xyxyxyxyxyx')
# Markers : matplotlib.markers

# you can save image in png or jpg format
plt.savefig('teeet.png')
```

> subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x*2,y,'b.-')
plt.subplot(2,2,4)
plt.plot(x,y,'r1')
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y,'r')
```

> Barplot

```python
x = [2,4,8,10,6]
y=[10,12,5,33,12]
x2 = [4,5,4,15,2]
y2 = [6,45,12,23,22]
```

```python
plt.bar(x,y,color = 'b')
plt.bar(x2,y2,color='g')
plt.show()
```

> Histogram

```python
a = np.random.randint(12,48,(450,1))
```

```python
plt.hist(a,bins=50)
plt.show()
```

> Boxplot

```python
data = [np.random.normal(0,std,100)for std in range(1,4)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

> Pie chart

```python
labels = 'python','C','Ruby'
sizes = [25,45,80]
colors = ['gold','red','blue']

explode = (0.1,0.2,0.04)
```

```python
#plot
plt.pie(sizes,explode = explode,labels =labels,colors =colors,
        autopct='%1.10f%%',shadow =True)
plt.axis('equal')
plt.show()
```
